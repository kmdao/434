Basic Probabilities

```python
pA = len(real1) / len(pulse) #calculating probability that a pulsar is class 1
print(pA)

mean80 = pulse[pulse[0] < 80] #separating pulsars with mean less than 80
pB = len(mean80) / len(pulse) #calculating probability that a pulsar has a mean less than 80
print(pB)

realMean80 = mean80[mean80[8] == 1] #separating class 1 pulsars that have mean less than 80
pAB = len(realMean80) / len(pulse) #calculating probability that a pulsar is both class 1 and has a mean less than 80
print(pAB)
```

Conditional Probabilities

P(A|B) = P(A $\cap$ B) / P(B)

P(B|A) = P(A|B) P(B) / P(A) = P(A $\cap$ B) / P(A)

```python
pAgivenB = pAB / pB #calculating probability that a pulsar is class 1 given that its mean is less than 80
print(pAgivenB)

pBgivenA = pAB / pA #calculating probability that a pulsar's mean is less than 80 given that it is class 1
print(pBgivenA)
```

Cumulative Probabilities

```python
gauss0 = norm.cdf(0) #calculating cumulative probabilities for Z = 0, 1, 3, and 5
gauss1 = norm.cdf(1)
gauss3 = norm.cdf(3)
gauss5 = norm.cdf(5)

mean1 = gauss1 - gauss0 #calculating probabilities from mean for Z = 1, 3, and 5
mean3 = gauss3 - gauss0
mean5 = gauss5 - gauss0

comp1 = 1 - gauss1 #calculating complementary probabilities for Z = 1, 3, and 5
comp3 = 1 - gauss3
comp5 = 1 - gauss5
```

```python
print('Cumulative probability from mean for Z = 1:')
print(mean1)
print('Cumulative probability from mean for Z = 3:')
print(mean3)
print('Cumulative probability from mean for Z = 5:')
print(mean5)
```

```python
print('Cumulative probability for Z = 1:')
print(gauss1)
print('Cumulative probability for Z = 3:')
print(gauss3)
print('Cumulative probability for Z = 5:')
print(gauss5)
```

```python
print('Complementary probability for Z = 1:')
print(comp1)
print('Complementary probability for Z = 3:')
print(comp3)
print('Complementary probability for Z = 5:')
print(comp5)
```
